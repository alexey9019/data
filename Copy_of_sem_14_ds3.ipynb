#### Маркетинговая компания

Как финансовое учреждение может повысить эффективность будущих маркетинговых кампаний? Найдите лучшие стратегии для улучшения для следующей маркетинговой кампании. Чтобы ответить на этот вопрос, мы должны проанализировать последнюю маркетинговую кампанию, проведенную банком, и выявить закономерности, которые помогут нам сделать выводы для разработки будущих стратегий.

Процесс, с помощью которого компании создают ценность для клиентов и выстраивают прочные отношения с клиентами, чтобы получать прибыль от клиентов в ответ. 

Маркетинговые кампании характеризуются ориентацией на потребности клиентов и их общую удовлетворенность. Тем не менее, существуют разные переменные, которые определяют, будет ли маркетинговая кампания успешной или нет. 

Существуют определенные переменные, которые мы должны учитывать при проведении маркетинговой кампании. 

* Сегмент населения: к какой части населения относится маркетинговая кампания и почему? Этот аспект маркетинговой кампании чрезвычайно важен, поскольку он определяет, какая часть населения, скорее всего, должна получить сообщение о маркетинговой кампании. 
* Канал дистрибуции, чтобы добраться до места клиента: внедрение наиболее эффективной стратегии, чтобы получить максимальную отдачу от этой маркетинговой кампании. К какой части населения мы должны обратиться? Какой инструмент мы должны использовать, чтобы получить наше сообщение? (Например: телефоны, радио, телевидение, социальные сети и т. Д.) 
* Цена: Какую цену лучше всего предложить потенциальным клиентам? (В случае маркетинговой кампании банка в этом нет необходимости, поскольку основной интерес для банка заключается в том, чтобы потенциальные клиенты открывали депозитарные счета для обеспечения продолжения оперативной деятельности банка.) 
* Стратегия продвижения: это как стратегия будет реализована и как будут рассматриваться потенциальные клиенты. Это должно быть последней частью анализа маркетинговой кампании, поскольку необходимо провести углубленный анализ предыдущих кампаний (если это возможно), чтобы извлечь уроки из предыдущих ошибок и определить, как сделать маркетинговую кампанию намного более эффективной.

#### Описание данных

данные клиента банка:
1 - возраст: (числовой)
2 - работа: тип работы (категориальная: «админ», «рабочий», «предприниматель», «домработница», «менеджмент», «пенсионер», «работающий не по найму», «услуги», «студент») 'техник', 'безработный', 'неизвестно')
3 - семейное положение: семейное положение (категориальное: «в разводе», «замужем», «не замужем», «неизвестно»; примечание: «в разводе» означает в разводе или вдове)
4 - образование: (категориальное: начальное, среднее, высшее и неизвестное)
5 - по умолчанию: есть кредит по умолчанию? (категорически: «нет», «да», «неизвестно»)
6 - жилье: есть жилищный кредит? (категорически: «нет», «да», «неизвестно»)
7 - кредит: есть личный кредит? (категорически: «нет», «да», «неизвестно»)
8 - баланс: баланс человека.

Связано с последним контактом текущей кампании:

8 - контакт: тип контактной связи (категориальный: «сотовый», «телефонный»)
9 - месяц: последний контактный месяц года (категориальный: «январь», «февраль», «мар», ..., «ноябрь», «декабрь»)
10-й день: последний контактный день недели (категориальный: понедельник, вторник, четверг, четверг, пятница)
11 - длительность: длительность последнего контакта в секундах (числовая). Важное примечание: этот атрибут сильно влияет на выходную цель (например, если длительность = 0, тогда y = 'нет'). Тем не менее, продолжительность не известна до выполнения вызова. Также после окончания звонка у очевидно известно. Таким образом, этот вход должен быть включен только для целей сравнения и должен быть отброшен, если предполагается, что у вас будет реалистичная модель прогнозирования.

другие атрибуты:

12 - кампания: количество контактов, выполненных во время этой кампании и для этого клиента (числовое, включая последний контакт)
13 - pdays: количество дней, прошедших после последнего контакта с клиентом из предыдущей кампании (числовое значение; 999 означает, что с клиентом ранее не связывались)
14 - предыдущая: количество контактов, выполненных до этой кампании и для этого клиента (числовое)
15 - результат: результат предыдущей маркетинговой кампании (категориальный: «провал», «несуществующий», «успех»)

Выходная переменная (желаемая цель):
21-й - подписал ли клиент срочный депозит? (бинарный: «да», «нет»)

```
# Подключаем класс для предобработки данных
from sklearn import preprocessing

# Напишем функцию, которая принимает на вход DataFrame, кодирует числовыми значениями категориальные признаки
# и возвращает обновленный DataFrame и сами кодировщики.
def number_encode_features(init_df):
    result = init_df.copy() # копируем нашу исходную таблицу
    encoders = {}
    for column in result.columns:
        if result.dtypes[column] == np.object: # np.object -- строковый тип / если тип столбца - строка, то нужно его закодировать
            encoders[column] = preprocessing.LabelEncoder() # для колонки column создаем кодировщик
            result[column] = encoders[column].fit_transform(result[column]) # применяем кодировщик к столбцу и перезаписываем столбец
    return result, encoders

encoded_data, encoders = number_encode_features(df) # Теперь encoded data содержит закодированные кат. признаки 
encoded_data.head() 
```

## Масштабирование признаков

### One hot

```
# пример работы функции get_dummies из pandas
# создали произвольный датасет
df_sample = pd.DataFrame({'A': ['a', 'b', 'a'], 'B': ['b', 'a', 'c'], 'C': [1, 2, 3]})
df_sample.head()
```

```
# результаты выбора определенный колонок
pd.get_dummies(df_sample, prefix=['col1', 'col2'])
# получили формат название выбранных колонок (в prefix) + _ + название уникального признака в выбранной колонке
# это one hot кодирование датасета
```

```
# находим колонки с типом данных объект (строка) и берем их название через .columns
# конвертируем в список и записываем в переменную cat_columns
cat_columns = list(df.select_dtypes(include='object').columns)
# отображаем датасет с выбранными категориальными
df[cat_columns].head()
```

```
# one-hot кодировали категориальные колонки датафрейма
dummies_df = pd.get_dummies(df, prefix=cat_columns)
dummies_df.head()
```

```
# название колонок итоговой обучающей выборки и записали в переменную dummies_columns
dummies_columns = list(dummies_df.columns)
print(dummies_columns)
# преобразовали колонки в numpy
X = np.array(dummies_df, float)
# отобразили первый элемент обучающей выборки и его ответ
X[0], y[0]
```

Масштабирование признаков можно выполнить, например, одним из следующих способов способами:
 - $x_{new} = \dfrac{x - \mu}{\sigma}$, где $\mu, \sigma$ — среднее и стандартное отклонение значения признака по всей выборке (см. функцию [scale](http://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.scale.html))
 - $x_{new} = \dfrac{x - x_{min}}{x_{max} - x_{min}}$, где $[x_{min}, x_{max}]$ — минимальный интервал значений признака

```
# нормализуем наши данные
from sklearn.preprocessing import scale
# скейлим (первой формулой выше)
X_one_hot = scale(X, with_std=True, with_mean=True)
X_one_hot[0]
```

## KNN

### KNN - One hot

```
# Будем использовать модель k ближайших соседей, которую проходили на 5 занятии

# Инициализируем алгоритм
knn = KNeighborsClassifier()

# В sklearn есть специальный модуль для работы с кросс-валидацией
from sklearn.model_selection import GridSearchCV

# Зададим сетку - среди каких значений выбирать наилучший параметр.
knn_grid = {'n_neighbors': np.array(np.linspace(1, 100, 10), dtype='int')} # перебираем по параметру <<n_neighbors>>, по сетке заданной np.linspace(2, 100, 10)

# Создаем объект кросс-валидации
gs = GridSearchCV(knn, knn_grid, cv=5)

# Обучаем его
gs.fit(X_scaled, y)
```

```
# Функция отрисовки графиков
def grid_plot(x, y, x_label, title, y_label='cross_val, accuracy'):
    # определили размер графика
    plt.figure(figsize=(12, 6))
    # добавили сетку на фон
    plt.grid(True)
    # построили по х - число соседей, по y - точность
    plt.plot(x, y, 'go-')
    # добавили подписи осей и название графика
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(title)
```

```
# Строим график зависимости качества от числа соседей
# замечание: результаты обучения хранятся в атрибуте cv_results_ объекта gs
grid_plot(knn_grid['n_neighbors'], gs.cv_results_['mean_test_score'], 'n_neighbors', 'KNeighborsClassifier')
```

Получаем вывод, что наиболее оптимальные параметры лежат +- на уровне 20-25 соседей, исследуем детальнее в этом диапазоне

```
# Зададим сетку - среди каких значений выбирать наилучший параметр.
knn_grid = {'n_neighbors': list(range(40, 50))} # перебираем по параметру <<n_neighbors>>, по сетке заданной np.linspace(2, 100, 10)

# Создаем объект кросс-валидации
gs = GridSearchCV(knn, knn_grid, cv=5)

# Обучаем его
gs.fit(X_scaled, y)
```

```
grid_plot(knn_grid['n_neighbors'], gs.cv_results_['mean_test_score'], 'n_neighbors', 'KNeighborsClassifier')
```

```
# Из графика видим, что самое оптимальное число соседей - 41, отобразим лучший параметр
gs.best_params_, gs.best_score_
```

```
def plot_feature_importances(model, columns):
    # выбираем кол-во признаков для отображения
    nr_f = 10
    # берем данные для графика
    # берем алгоритм с лучшими параметрами
    # по х мы берем важность признаков исходя из критерия Джини, сортируем их вместе с названиями колонок по возрастанию
    imp = pd.Series(data = model.best_estimator_.feature_importances_, 
                    index=columns).sort_values(ascending=False)
    # построили фигуру
    plt.figure(figsize=(7,5))
    # отобразили название
    plt.title("Важность признаков | Feature importance")
    # построили 10 самых важных признаков
    ax = sns.barplot(y=imp.index[:nr_f], x=imp.values[:nr_f], orient='h')
```
